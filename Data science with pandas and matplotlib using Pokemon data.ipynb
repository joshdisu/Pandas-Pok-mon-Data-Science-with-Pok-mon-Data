```python
columns = [
    "english_name",
    "hp",
    "attack",
    "defense",
    "sp_attack",
    "sp_defense",
    "speed",
    "primary_type",
    "is_sublegendary",
    "is_legendary",
    "is_mythical",
]
pokemon = df[columns]

pokemon.head()
```

```python
pokemon
```

```python
#stat formulas with broadcasting 

stats = [
    "hp",
    "attack",
    "defense",
    "sp_attack",
    "sp_defense",
    "speed",
]
```

```python
2 * pokemon[stats[1:]] + 5
```

```python
#Use broadcasting to compute final values of six stats 
2 * pokemon[stats] + (110, 5, 5, 5, 5, 5)
```

```python
#Save stats into dataframe 

pokemon.join(
    2 * pokemon[stats] + (110, 5, 5, 5, 5, 5),
    rsuffix = "_final",
)
```

```python
#Total stat for each Pokemon 

pokemon[stats].sum(axis=1)
```

```python
pokemon = pokemon.assign(
   total = pokemon[stats].sum(axis=1)
)
pokemon.head()
```

```python
pokemon = pokemon.astype(
    {
        "is_mythical": "bool",
        "is_legendary": "bool",
        "is_sublegendary": "bool"
    }
)

```

```python
pokemon = pokemon.assign(
    legendary = pokemon[ ["is_mythical", "is_legendary", "is_sublegendary"] ].any(axis=1)
)
```

```python
pokemon[ pokemon["legendary"] ]
```

```python
#Convert column variable to categorical values 

pokemon = pokemon.astype(
    {"primary_type": "category"}
)
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()
```

```python
fig, ax = plt.subplots(
    figsize=(4, 2.25),
    facecolor="whitesmoke",
    layout="constrained",
)

ax.plot(pokemon["hp"])
ax.set_title("HP of all Pokemon")
ax.set_xlabel("Pokemon index")
ax.set_ylabel("Base HP")
```

```python
pokemon["primary_type"].value_counts()
```

```python
fig, ax = plt.subplots(
    figsize=(4, 2.25),
    facecolor="whitesmoke",
    layout="constrained",
)

primary_type = pokemon["primary_type"]

ax.bar(
 range(primary_type.nunique()),
 primary_type.value_counts(),
 tick_label=primary_type.cat.categories
)
ax.set_title("Distribution of Pokemon per type")
ax.set_xlabel("Primary type")
ax.set_ylabel("Count")
```

```python
primary_type.cat.categories
```

```python
fig, ax = plt.subplots(
    figsize=(8, 5),
    facecolor="whitesmoke",
    layout="constrained",
)

primary_type = pokemon["primary_type"]
value_counts = primary_type.value_counts()
colours = [type_colours_data[ptype] for ptype in value_counts.index]

ax.barh(
 range(primary_type.nunique()),
 primary_type.value_counts(),
 color=colours,
)
ax.set_yticks(
    range(primary_type.nunique()),
    labels=primary_type.value_counts().index,

)
ax.set_title("Distribution of Pokemon per type")
ax.set_ylabel("Primary type")
ax.set_xlabel("Count")

fig.savefig("best_plot.png")
```

```python
#Categorise bar columns by colour 

type_colours_data = {
    "normal": "#A8A77A",
    "fire": "#EE8130",
    "water": "#6390F0",
    "electric": "#F7D02C",
    "grass": "#7AC74C",
    "ice": "#96D9D6",
    "fighting": "#C22E28",
    "poison": "#A33EA1",
    "ground": "#E2BF65",
    "flying": "#A98FF3",
    "psychic": "#F95587",
    "bug": "#A6B91A",
    "rock": "#B6A136",
    "ghost": "#735797",
    "dragon": "#6F35FC",
    "dark": "#705746",
    "steel": "#B7B7CE",
    "fairy": "#D685AD",
}
```

```python
fig, ax = plt.subplots(
    figsize=(8, 5),
    facecolor="whitesmoke",
    layout="constrained",
)

ax.violinplot(
    [
    pokemon["speed"][ pokemon["primary_type"] == "electric"],
    pokemon["speed"][ pokemon["primary_type"] != "electric"],
    pokemon["speed"][pokemon["primary_type"] == "flying"],
    pokemon["speed"][pokemon["primary_type"] != "electric"],
    ],
    showmeans=True
)

```
